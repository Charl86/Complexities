```python
PRIME = 5
POWER = 2
```

```python
gfp = GF(PRIME^POWER, "a")
gfp
```

```python
class Sidelnikov:
    def __init__(self, q):
        self.q = q
        
        self.galoisFieldq = GF(self.q, 'a')
        
        if not is_prime_power(self.q):
            raise TypeError(f"{self.q} must be a prime or a prime power.")
    
    def __call__(self, t):
        primitive_elm = self.galoisFieldq.primitive_element()

        if not (primitive_elm^t + 1).is_square():
            return 1
        else:
            return 0


class Dummy:
    def __init__(self, q):
        self.q = q
    
    def __call__(self, i):
        return i % (self.q - 1)
```

```python
s = Sidelnikov(PRIME^POWER)
# s = Dummy(PRIME^POWER)

# Applying s(t) on elements of GF(PRIME^POWER)
for t in range(len(gfp)):
    print(f"s({t}) = {s(t)}")
```

```python
T = matrix(ZZ, PRIME, PRIME)
```

```python
for elm in gfp:
    if elm != 0:
        elmCoeffs = elm.polynomial().list()
        
        if len(elmCoeffs) == 1:
            x = 0
        else:
            x = elmCoeffs[1]
        y = elmCoeffs[0]
        
        power = elm.log(gfp.primitive_element())

        T.add_to_entry(y, x, power)
    else:
        T.add_to_entry(0, 0, -1)
        
T.matrix_from_rows(reversed(range(PRIME)))
```

```python
A = list()
for elm in gfp:
    elmCoeffs = elm.polynomial().list()

    if elm != 0:
        if len(elmCoeffs) == 1:
            i = 0
        else:
            i = elmCoeffs[1]
        j = elmCoeffs[0]
    else:
        i, j = 0, 0

    for k in range(PRIME^POWER - 1):
        if i == 0 and j == 0:
            # Convert to 0 when passing to executable
            entry = 0
        else:
            entry = s((k - T[j % PRIME][i % PRIME]) % (PRIME^POWER - 1))

        coordentry = [(i,j,k), entry]
        print(f"{coordentry[0]}: {entry}")
        A.append(coordentry)
```
